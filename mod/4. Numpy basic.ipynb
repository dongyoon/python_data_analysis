### 4.1.1. ndarray 생성

```python
data1 = [6,7.5,8,0,1] ## data1에 리스트(element가 5개) 저장 
arr1 = np.array(data1) ## np.array로 리스트를 ndarray로 변환
print(arr1)
print(type(arr1)) ## arr1은 ndarray 타입
```

```python
data2 = [[1,2,3,4],[5,6,7,8]] ## 리스트 내에 리스트를 포함
arr2 = np.array(data2) ## data2를 2차원 배열로 변환
print(arr2) 
print(arr2.ndim) ## arr2의 디멘젼 갯수 - 2차원 배열
print(arr2.shape) ## arr2의 디멘젼
```

```python
print(np.zeros(10)) ## 디멘젼에 해당하는 배열 생성 후, 0 대입
print(np.zeros((3,6)))
```

```python
print(np.empty((2,3,2))) ## 3차원의 배열 생성 후, 0이 아닌 값 대입
print(np.arange(15)) ## 0부터 15개의 연속적인 정수로 구성된 1차원 배열 생성 
```

### 4.1.2. ndarray의 자료형

```python
arr1 = np.array([1,2,3], dtype = np.float64)
arr2 = np.array([1,2,3], dtype = np.int32)
print(arr1.dtype)
print(arr2.dtype)
```

```python
arr = np.array([1,2,3,4,5])
print("int type : ", arr.dtype)
arr = arr.astype(np.float64) ## 타입을 플로트로 변경
print("float type : ", arr.dtype)
```

```python
arr = np.array([3.7,2.8,3.1,4.2,5.4])
print(arr)
print("float type : ", arr.dtype)
arr = arr.astype(np.int32) ## 타입을 정수로 변경, 문자열도 숫자로 변환 가능
print(arr) ## 소수점을 버림
print("int type : ", arr.dtype)
```

### 4.1.5. 불리언 색인

```python
names = np.array(['Bob', "J", "W", "Bob","W", "J","J"])
data = np.random.randn(7,4)
print(names)
print(data)
```

```python
print(names == "Bob") ## 0,3번쨰 element가 True로 리턴
print(data[names == "Bob"]) ## 0, 3번째 row만 리턴
```

```python
print(data[names == "Bob", 2:]) ## 0, 3번째 row & 2,3번째 column만 리턴
```

```python
print(data[-(names == "Bob"), 2:]) ## 앞에 -를 붙이면 해당 row제외 --> Bob이 아닌 1,2,4,5,6번째 row & 2, 3번째 column만 리턴
```

```python
print(data)
data[data<0] = 0 ## 0보다 작은 element에 0 대입
print(data)
```

### 4.1.6. Fancy indexing

```python
arr = np.empty((8,4))
print(arr)
for i in range(8):
    arr[i] = i
print(arr)
```

```python
print(arr[[4,3,0,6]]) ## 4, 3, 0, 6 순으로 row를 리턴
print(arr[[-3,-5,-7]]) ## 끝에서 3, 5, 7 번쨰 row를 리턴
```

```python
arr = np.arange(32).reshape(8,4)
print(arr)
```

```python
print(arr[[1,5,7,2],[0,3,1,2]]) ## (1,0), (5,3)... element returns
```

```python
print(arr[[1,5,7,2]]) ## select 1,5,7,2 rows
print(arr[[1,5,7,2]][:,[0,3,1]]) ## select 1,5,7,2 rows + select 0,3,1 columns 
print(arr[np.ix_([1,5,7,2],[0,3,1])]) ## same as the above 
```

### 4.1.7. Transporsing arrays and swapping axes

```python
arr = np.arange(15).reshape([3,5])
print(arr)
print(arr.T)
print(arr.transpose(1,0)) ## .T is the special form of transpose
print(arr.transpose(0,1))
print(arr.swapaxes(0,1)) ## .T is the special form of swapaxes
print(np.dot(arr, arr.T)) ## dot product of arr array 
```

### 4.2.2. Expressing conditional logic as array operations

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True,False,True,True,False])
result = [(x if c else y) for x,y,c in zip(xarr,yarr,cond)] ## 1) takes long time cause don't use numpy
print(result) ## 2) will not work with multidimensional arrays, why has different values?
```

```python
result = np.where(cond, xarr, yarr) ## with np.where it can be represented more concisely
print(result)
```

```python
arr = np.random.randn(4,4)
print(arr)
print(np.where(arr>0,2,-2))
print(np.where(arr>0,2,arr))
```

## 4.4. Linear algebra

```python
x = np.array([[1.,2.,3.,],[4.,5.,6.]])
y = np.array([[6.,23.],[-1,7],[8,9]])
print(x)
print(y)
```

```python
print(x.dot(y)) ## dot product of two matrix
print(np.dot(x,y))
```

```python
np.dot(x, np.ones(3))
```

```python
from numpy.linalg import inv, qr
x = np.random.randn(5,5)
mat = x.T.dot(x)
```

```python
print(mat)
print(inv(mat))
print(np.dot(mat, inv(mat)))
print(mat.dot(inv(mat)))
```

```python
q, r= qr(mat) ## QR decomposition
print(r)
```
